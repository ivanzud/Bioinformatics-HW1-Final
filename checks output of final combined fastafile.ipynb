```python
seq_list =[]
output_file = "inputfile.fasta"
ofile1 = open(output_file, "r")
ofile_content1 = ofile1.read()
#print(ofile_content)
words1 = str.split(ofile_content1)
#print(words)
for i in range(len(words1)):
    if("S" not in words1[i]):
        #print("hello")
        seq_list.append(words1[i])
ofile1.close()
```
